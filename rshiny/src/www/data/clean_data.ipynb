```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/owid-covid-data.csv")
```

```python
df.head()
```

```python
df = df[["location","people_vaccinated","date","people_fully_vaccinated","population"]]
```

```python
rslt_df = df[df['date'] > "2022-01-01"]
```

```python
rslt_df
```

```python
df1 = rslt_df.dropna()
```

```python
df1
```

```python
g = df1.groupby('location')
g.tail(1)
```

```python
df2 = g.tail(1)
```

```python
df2
```

```python
df3 = df2[["location","people_vaccinated","people_fully_vaccinated","population"]]
```

```python
df3["first"] = df3["people_vaccinated"]/df3["population"]
df3["fully"] = df3["people_fully_vaccinated"]/df3["population"]
```

```python
df3 = df3[["location","first","fully"]]
```

```python
df3.rename(columns={'location':'country'}, inplace=True)
df3 = df3.reset_index()
df3 = df3[["country","first","fully"]]
```

```python
df3
```
