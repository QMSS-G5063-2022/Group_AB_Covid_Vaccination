```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("owid-covid-data .csv")
```

```python
df.head()
```

```python
df = df[["location","total_cases","total_deaths","total_vaccinations","date","population"]]
```

```python
df_date = df[df['date'] > "2022-01-01"]
df_date.head()
```

```python
df_date = df_date.dropna()
df_date
```

```python
df_date["infection_rate"] = df_date["total_cases"]/df_date["population"]
df_date["death_rate"] = df_date["total_deaths"]/df_date["population"]
df_date["vaccines_offered"] = df_date["total_vaccinations"]
```

```python
df_date
```

```python
df1 = df_date[["location","infection_rate","death_rate","vaccines_offered","date"]]
```
